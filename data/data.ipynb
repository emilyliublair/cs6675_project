```python
import os
import matplotlib.pyplot as plt
import numpy as np

folder_path = "./all"
file_sizes = []

# Gather sizes of all .txt files
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        filepath = os.path.join(folder_path, filename)
        file_size = os.path.getsize(filepath)  # Size in bytes
        file_sizes.append(file_size)

# Convert to NumPy array for easy stats
file_sizes_np = np.array(file_sizes)

# Compute statistics
count = len(file_sizes_np)
mean = np.mean(file_sizes_np)
median = np.median(file_sizes_np)
minimum = np.min(file_sizes_np)
maximum = np.max(file_sizes_np)
std_dev = np.std(file_sizes_np)

# Print statistics
print(f"Number of .txt files: {count}")
print(f"Mean size: {mean:.2f} bytes")
print(f"Median size: {median:.2f} bytes")
print(f"Min size: {minimum} bytes")
print(f"Max size: {maximum} bytes")
print(f"Standard Deviation: {std_dev:.2f} bytes")

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(file_sizes, bins=30, edgecolor='black')
plt.title("Distribution of File Sizes")
plt.xlabel("File size (bytes)")
plt.ylabel("Number of files")
plt.grid(True)
plt.tight_layout()
plt.show()
```
